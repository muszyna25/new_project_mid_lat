```python
# System imports
from __future__ import print_function
from __future__ import division
from __future__ import absolute_import
import os

# External imports
import ipyparallel as ipp
import numpy as np
import matplotlib.pyplot as plt
import glob
from sklearn import metrics
import random
import csv
import glob
import matplotlib.pyplot as plt 

%matplotlib notebook
%matplotlib inline
```

```python
def read_results_from_csv(fname):
    input_file = csv.DictReader(open(fname))
    my_list = []
    for row in input_file:
        my_list.append(row)
    return my_list

def plot_imgs(lr, tit):
    plt.figure()
    plt.gca().invert_yaxis()
    #plt.yscale('log',basey=10)
    print(lr)
    plt.plot(range(0,len(lr)), lr)
    plt.title(tit)
    plt.grid()
    plt.show()
    
def plot_acc(train, val, tit):
    fig, ax = plt.subplots()
    print(lr)
    train=[round(float(x),2) for x in train]
    val=[round(float(x),2) for x in val]
    ax.plot(train, label='Train ACC')
    ax.plot(val, label='Val ACC')
    #plt.title(tit)
    ax.grid()
    plt.legend(loc='lower right')
    plt.show()

def plot_loss(train, val, tit):
    fig, ax = plt.subplots()
    print(lr)
    train=[round(float(x),2) for x in train]
    val=[round(float(x),2) for x in val]
    ax.plot(train, label='Train Loss')
    ax.plot(val, label='Val Loss')
    #plt.title(tit)
    ax.grid()
    plt.legend(loc='upper right')
    plt.show()
    
def plot_all_measures(l_train, l_val, train_acc, val_acc, lr):
    fig, axs = plt.subplots(3,1, figsize=(18, 12), facecolor='w', edgecolor='k')
    fig.subplots_adjust(hspace = 0.5, wspace=.01)
    axs = axs.ravel()
    
    l_train=[round(float(x),2) for x in l_train]
    l_val=[round(float(x),2) for x in l_val]
    axs[0].plot(l_train, label='Train Loss')
    axs[0].plot(l_val, label='Val Loss')
    axs[0].legend(loc='upper right')
    axs[0].grid()
    
    train_acc=[round(float(x),2) for x in train_acc]
    val_acc=[round(float(x),2) for x in val_acc]
    axs[1].plot(train_acc, label='Train ACC')
    axs[1].plot(val_acc, label='Val ACC')
    axs[1].legend(loc='lower right')
    axs[1].grid()
    
    plt.gca().invert_yaxis()
    axs[2].plot(lr, label='Learning Rate')
    axs[2].legend(loc='upper right')
    axs[2].grid()
    plt.show()
```

```python
i_model = '24977547'
dir_path = '/global/cscratch1/sd/muszyng/cnn_model_runs/Model_0/' + i_model + '/'
```

```python
lf = glob.glob(dir_path + 'training_*.log')
lf
```

```python
#epoch,acc,f1_m,loss,lr,precision_m,recall_m,val_acc,val_f1_m,val_loss,val_precision_m,val_recall_m
l_train_acc = []
l_val_acc = []
l_val_precision_m = []
l_val_recall_m = []
l_lrate = []

train_acc = []
val_acc = []
loss_train = []
loss_val = []

for fn in lf:
    data = read_results_from_csv(fn)
    last_train_acc = data[len(data)-1]['acc']
    last_val_acc = data[len(data)-1]['val_acc']

    indx = fn.split('_')[4].split('.')[0]

    train_acc.append((indx, [data[i]['acc'] for i in range(0, len(data))]))
    val_acc.append((indx, [data[i]['val_acc'] for i in range(0, len(data))]))
    loss_train.append((indx, [data[i]['loss'] for i in range(0, len(data))]))
    loss_val.append((indx, [data[i]['val_loss'] for i in range(0, len(data))]))
    
    l_train_acc.append((indx, last_train_acc, last_val_acc))
    l_lrate.append((indx, [data[i]['lr'] for i in range(0, len(data))]))
```

```python
best_train_acc = sorted(l_train_acc, key=lambda x: x[1], reverse=True)[:10]
worst_train_acc = sorted(l_train_acc, key=lambda x: x[1])[:10]
print('Best: ', *best_train_acc, sep='\n')
print('Worst: ', *worst_train_acc, sep='\n')
```
